Separação das categorias de Importaçâo e exportação de valor

```python
#Fazendo o tratamento da base de exportação
base = ['Mês', 'Países', 'Código ISIC Classe', 'Descrição ISIC Classe', 'UF do Produto', 'Via', 'Código NCM', 'Descrição NCM']
export= [x for x in  comex if 'Exportação' in x and  'Valor' in x] 
exportacao_usd = comex[base+export]
exportacao_usd = exportacao_usd.melt(id_vars= base)
exportacao_usd['Mês'] = exportacao_usd['Mês'].apply(lambda x :(x.split('.')[1])).str.rstrip(' ')
def mes(x):
    if 'Janeiro' in x:
        return 1
    elif 'Fevereiro' in x:
        return 2
    elif 'Março' in x:
        return 3
    elif 'Abril' in x:
        return 4
    elif 'Maio' in x:
        return 5
    elif 'Junho' in x:
        return 6
    elif 'Julho' in x:
        return 7
    elif 'Agosto' in x:
        return 8
    elif 'Setembro' in x:
        return 9
    elif 'Outubro' in x:
        return 10 
    elif 'Novembro' in x:
        return 11
    elif 'Dezembro' in x:
        return 12
exportacao_usd['Mes_num'] = exportacao_usd['Mês'].apply(mes)
exportacao_usd[['Tipo', 'Ano', 'Indicador']] = exportacao_usd['variable'].str.split( ' - ',n=2,expand= True)
exportacao_usd = exportacao_usd.drop(columns= {'variable'})
exportacao_usd['Via'] = exportacao_usd['Via'].str.lower().str.capitalize() #ou .title()

exportacao_usd
```

```python
#Fazendo o tratamento da base de importação
base = ['Mês', 'Países', 'Código ISIC Classe', 'Descrição ISIC Classe', 'UF do Produto', 'Via', 'Código NCM', 'Descrição NCM']
importa = [x for x in  comex if 'Importação' in x and 'Valor' in x] 
importacao_usd = comex[base+importa]
importacao_usd = importacao_usd.melt(id_vars= base)
importacao_usd['Mês'] = importacao_usd['Mês'].apply(lambda x :(x.split('.')[1])).str.rstrip(' ')
def mes(x):
    if 'Janeiro' in x:
        return 1
    elif 'Fevereiro' in x:
        return 2
    elif 'Março' in x:
        return 3
    elif 'Abril' in x:
        return 4
    elif 'Maio' in x:
        return 5
    elif 'Junho' in x:
        return 6
    elif 'Julho' in x:
        return 7
    elif 'Agosto' in x:
        return 8
    elif 'Setembro' in x:
        return 9
    elif 'Outubro' in x:
        return 10 
    elif 'Novembro' in x:
        return 11
    elif 'Dezembro' in x:
        return 12
importacao_usd['Mes_num'] = importacao_usd['Mês'].apply(mes)
importacao_usd[['Tipo', 'Ano', 'Indicador']] = importacao_usd['variable'].str.split( ' - ',n=2,expand= True)
importacao_usd = importacao_usd.drop(columns= {'variable'})
importacao_usd['Via'] = importacao_usd['Via'].str.lower().str.capitalize() #ou .title()
importacao_usd
```

```python
#Concatenando as duas partes
Total_valor = pd.concat([exportacao_usd,importacao_usd]).drop(columns={'Indicador'})
Total_valor.rename(columns={'value': 'US$'},inplace= True)
Total_valor
```

Separação das categorias de Importaçâo e exportação de Quilograma

```python
#Fazendo o tratamento da base de exportação
base = ['Mês', 'Países', 'Código ISIC Classe', 'Descrição ISIC Classe', 'UF do Produto', 'Via', 'Código NCM', 'Descrição NCM']
export= [x for x in  comex if 'Exportação' in x and  'Quilograma' in x] 
exportacao = comex[base+export]
exportacao = exportacao.melt(id_vars= base)
exportacao['Mês'] = exportacao['Mês'].apply(lambda x :(x.split('.')[1])).str.rstrip(' ')
def mes(x):
    if 'Janeiro' in x:
        return 1
    elif 'Fevereiro' in x:
        return 2
    elif 'Março' in x:
        return 3
    elif 'Abril' in x:
        return 4
    elif 'Maio' in x:
        return 5
    elif 'Junho' in x:
        return 6
    elif 'Julho' in x:
        return 7
    elif 'Agosto' in x:
        return 8
    elif 'Setembro' in x:
        return 9
    elif 'Outubro' in x:
        return 10 
    elif 'Novembro' in x:
        return 11
    elif 'Dezembro' in x:
        return 12
exportacao['Mes_num'] = exportacao['Mês'].apply(mes)
exportacao[['Tipo', 'Ano', 'Indicador']] = exportacao['variable'].str.split( ' - ',n=2,expand= True)
exportacao = exportacao.drop(columns= {'variable'})
exportacao['Via'] = exportacao['Via'].str.lower().str.capitalize() #ou .title()

exportacao
```

```python
#Fazendo o tratamento da base de importação
base = ['Mês', 'Países', 'Código ISIC Classe', 'Descrição ISIC Classe', 'UF do Produto', 'Via', 'Código NCM', 'Descrição NCM']
importa = [x for x in  comex if 'Importação' in x and 'Quilograma' in x] 
importacao = comex[base+importa]
importacao = importacao.melt(id_vars= base)
importacao['Mês'] = importacao['Mês'].apply(lambda x :(x.split('.')[1])).str.rstrip(' ')
def mes(x):
    if 'Janeiro' in x:
        return 1
    elif 'Fevereiro' in x:
        return 2
    elif 'Março' in x:
        return 3
    elif 'Abril' in x:
        return 4
    elif 'Maio' in x:
        return 5
    elif 'Junho' in x:
        return 6
    elif 'Julho' in x:
        return 7
    elif 'Agosto' in x:
        return 8
    elif 'Setembro' in x:
        return 9
    elif 'Outubro' in x:
        return 10 
    elif 'Novembro' in x:
        return 11
    elif 'Dezembro' in x:
        return 12
importacao['Mes_num'] = importacao['Mês'].apply(mes)
importacao[['Tipo', 'Ano', 'Indicador']] = importacao['variable'].str.split( ' - ',n=2,expand= True)
importacao = importacao.drop(columns= {'variable'})
importacao['Via'] = importacao['Via'].str.lower().str.capitalize() #ou .title()
importacao
```

```python
#Concatenando as duas partes
Total_Quilograma = pd.concat([exportacao,importacao]).drop(columns={'Indicador'})
Total_Quilograma.rename(columns={'value': 'KG'},inplace= True)
Total_Quilograma
```

Formando uma base

```python
#Unificando em uma base consolidada o tratamento de valor e quilograma
Base = pd.merge(Total_valor,Total_Quilograma, how= 'inner', on= ['Tipo', 'Ano', 'Código ISIC Classe', 'Mês', 'Descrição NCM', 'Código NCM', 'Mes_num', 'Via', 'Países', 'UF do Produto', 'Descrição ISIC Classe'])
colunas_ordem = ['Tipo', 'Ano', 'Mês', 'Mes_num','Código ISIC Classe', 'Descrição NCM', 'Código NCM', 'Via', 'Países', 'UF do Produto', 'Descrição ISIC Classe','KG','US$']
Base = Base[colunas_ordem]
Base
```
